Вариант 2. Базовый персептрон для логического ИЛИ (OR)

Реализуйте однослойный персептрон для логической функции OR. Используйте пороговую функцию активации (step function). Веса инициализируйте нулями. Обучите на всех 4 примерах. Выведите финальные веса и предсказания.

Импорты

```python
import numpy as np
```

Данные для логической функции OR:

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])
```

Веса, смещение и скорость обучения

```python
w = np.array([0.0, 0.0])
b = 0.0
learning_rate = 1.0
```

Пороговая функция активации:

```python
def step(x):
    return 1 if x >= 0 else 0
```

```python
epochs = 10
for ep in range(epochs):
    for i, (x1, x2) in enumerate(X):
        xn = w[0] * x1 + w[1] * x2 + b
        pred = step(xn)
        loss = y[i] - pred
        w[0] += learning_rate * loss * x1
        w[1] += learning_rate * loss * x2
        b += learning_rate * loss
print("Финальные веса: w1 = {:.2f}, w2 = {:.2f}, b = {:.2f}".format(w[0], w[1], b))

print("\nПредсказания для всех входов:")
for i in range(len(X)):
    net = np.dot(w, X[i]) + b
    pred = step(net)
    print(f"{X[i]} -> {pred} (ожидалось {y[i]})")
```
